```python
import requests
import json
import pandas as pd
```

```python
url = "http://127.0.0.1:5000/get-data?table=employee"
response = requests.request("GET",url)
employees = json.loads(response.text)
print(employees)
```

```python
url = "http://127.0.0.1:5000/get-data?table=office"
response = requests.request("GET",url)
offices = json.loads(response.text)
print(offices)
```

```python
url = "http://127.0.0.1:5000/get-data?table=check_in_out"
response = requests.request("GET",url)
check_in_out = json.loads(response.text)
print(check_in_out)
```

```python
employees_df = pd.DataFrame(employees)
employees_df.set_index('id',inplace=True)
employees_df
```

```python
offices_df = pd.DataFrame(offices)
offices_df.set_index('id',inplace=True)
offices_df
```

```python
check_in_out_df = pd.DataFrame(check_in_out)
check_in_out_df.set_index('id',inplace=True)
check_in_out_df.head()
```

```python
for index, row in check_in_out_df.iterrows():
    if(row['office_check_in_id']!=row['office_check_out_id']):
        check_in_out_df.at[index,'moved_offices'] = True
    else:
        check_in_out_df.at[index,'moved_offices'] = False
```

```python
check_in_out_df.head()
```

```python
type(check_in_out_df['check_in_date'][1])
```

```python
check_in_out_df['check_in_date'] = check_in_out_df['check_in_date'].str.replace(' GMT', '')
check_in_out_df['check_out_date'] = check_in_out_df['check_out_date'].str.replace(' GMT', '')
```

```python
from datetime import datetime
```

```python
datetime_format = "%a, %d %b %Y %H:%M:%S"
check_in_out_df['check_in_date'] = pd.to_datetime(check_in_out_df['check_in_date'], format=datetime_format)
check_in_out_df['check_out_date'] = pd.to_datetime(check_in_out_df['check_out_date'], format=datetime_format)
```

```python
type(check_in_out_df['check_in_date'][1])
```

```python
check_in_out_df['working_hours'] = (check_in_out_df['check_out_date'] - check_in_out_df['check_in_date']).dt.total_seconds() / 3600
check_in_out_df
```

```python
moved_offices_count = check_in_out_df['moved_offices'].value_counts()
moved_offices_count
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,6))
moved_offices_count.plot(kind='bar',color=['skyblue','salmon'])

# Adding labels and title
plt.xlabel('Moved Offices')
plt.ylabel('Employee Count')
plt.title('Employees Who Moved Offices vs. Those Who Didn\'t')

# Displaying the legend
plt.xticks(ticks=[0, 1], labels=['Did Not Move', 'Moved'], rotation=0)

# Show the plot
plt.tight_layout()
plt.show()
```
